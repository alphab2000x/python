```python
import seaborn as sns
import matplotlib.pyplot as mpl
import pandas as pd
```

```python
df = pd.read_csv('../data/IBM_HR_ATTRITION.csv') #Permet de lire un fichier .csv
```

```python
d = {'Yes': True, 'No': False} #On defini les noms des variables qui vont être transformer en bolean
df['Attrition'] = df['Attrition'].map(d) #On cast la colonne de notre tableau
```

```python
df['Attrition']=df.Attrition.replace('Yes',1)
df['Attrition']=df.Attrition.replace('No',0)
```

```python
g = sns.lmplot(x="DistanceFromHome", y="Attrition", data=df) #L'impact de la distance sur l'attrition
```
