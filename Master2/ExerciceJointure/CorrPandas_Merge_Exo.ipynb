## Importer pandas

```python
import pandas as pd
#import numpy as np
#import matplotlib as mpl
#import seaborn as sns
path = "data/"
```

## Importer les datasets & renommer la colonne 'Retail Branding' du dataset android_devices.csv

```python
ad = pd.read_csv(path + 'android_devices.csv')
ud = pd.read_csv(path + 'user_device.csv')
uu = pd.read_csv(path + 'user_usage.csv')


ad = ad.rename(columns={"Retail Branding": "manufacturer"});
```

## Votre premier merge

```python
result = uu.merge(ud, on='use_id', how='left').drop(columns=['user_id', 'platform_version', 'use_type_id'])
result.head()
```

## Afficher la shape de vos dataset ainsi que celle du dataset de sortie

```python
print("user_usage dimensions:  " , uu.shape)
print("user_device dimensions: " , ud.shape)
print("result dimensions:      " , result.shape)
```

##### On remarque que le dataset result (merge de user_device & user_usage) contient moins de lignes que les datasets d'origines.
##### Le merge prend l'intersection des deux dataframes donc ResultDim <= min(user_usageDim, user_deviceDim)

## Afficher via value_counts les use_id présent dans le nouveau dataset ainsi que ceux non présent

```python
use_id = result['device'].notna().rename("use_id")
use_id.value_counts()
#uu['use_id'].isin(ud['use_id']).value_counts()
```

## Le left merge

```python
print("user_usage dimensions: " , uu.shape)
print("result dimensions:     " , result.shape)
missing_values = result['device'].isnull().rename("use_id").value_counts()[True]
print("There are" , missing_values , "missing values in the result.")
```

## Afficher votre dataset

```python
result.head()
```

## Le right merge

```python
result = uu.merge(ud, on="use_id", how="right")
print("user_device dimensions: " , ud.shape)
print("result dimensions:      " , result.shape)
#missing_values = result['monthly_mb'].isnull().value_counts()[True]
missing_values = result["monthly_mb"].isna().sum()
print("There are" , missing_values , "missing values in the 'monthly_mb' column in the result.")
#missing_values = result.shape[False] - result['platform'].count()
missing_values = result["platform"].isna().sum()
print("There are" , missing_values , "missing values in the 'platform' column in the result.")
```

## Le outer merge example

```python
result = uu.merge(ud[['use_id', 'platform', 'device']], on="use_id", how="outer", indicator=True)
print("There are" , result.shape[0] , "unique values of use_id in our dataframes.")
print("Outer merge result has" , result.shape , "rows.")
print("There are", uu['use_id'].isin(ud['use_id']).value_counts()[True] , "rows with no missing values.")
```

## Ajouter les colonnes Model & manufacturer

```python
result = uu.merge(ud[['use_id', 'platform', 'device']], on="use_id", how="left")
#result.head()
result = result.merge(ad[['Device', 'manufacturer', 'Model']], left_on='device', right_on='Model', how='left')
result.head()
```
